```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib 
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
#list all columns 
all_data.columns
```

```python
#statistics sum
all_data.describe()
```

```python
all_data.shape
```

CLEAN UP DATA

```python
#drop nan value
all_data = all_data.dropna(how = 'all')
all_data.head()
```

```python
#drop or val
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
```

```python
#To convert all data in int 
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

```

ADD MONTH COLUMN

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

ADD SALES COLUMN 

```python
all_data['Sales'] = all_data['Quantity Ordered']* all_data['Price Each']
all_data.head()
```

ADD CITY COLUMN

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
all_data.head()
```

1.WHAT WAS THE BEST MONTH FOR SALES? HOW MUCH WAS EARNED THAT MONTH?

```python
all_data.groupby(['Month']).sum()['Sales']
```

```python
months = range(1,13)
print(months)
plt.title('Monthly Sales ')
plt.bar(months,all_data.groupby(['Month']).sum()['Sales'], color = 'g',edgecolor = 'b')
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

Maximum sales are seen in the month of december

3.WHAT TIME SHOULD WE DISPLAY ADVERTISEMENTS TO MAXIMIZE LIKELIHOOD OF CUSTOMERS BUYING THE PRODUCT ?

```python
# Add hour column
all_data['Hour'] = pd.to_datetime(all_data['Order Date']).dt.hour
all_data['Minute'] = pd.to_datetime(all_data['Order Date']).dt.minute
all_data['Count']=1
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby(['Hour'])]

plt.plot(hours, all_data.groupby(['Hour']).count(),'g:')
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.show()
```
